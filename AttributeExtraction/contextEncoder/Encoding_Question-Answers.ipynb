# Generating training data to trian model

```python
def generate_data(original_file,final_file):
    data_word_list = []
    data = [] 
    new_data = []
    predicates = []
    word_list = [] 
    with open(original_file, "r") as f:
            data_word_list = f.read().split("\n")
    for i in range(len(data_word_list)):
        #new_list.append(word_list[i].split(" "))
        data.append(tokenizer.tokenize(data_word_list[i]))

    with open(final_file, "r") as f:
            word_list = f.read().split("\n")
    for i in range(len(word_list)):
        new_data.append(word_list[i].split("\t"))

    for i in range(len(new_data)):
        if new_data[i][0].isnumeric():
            k =2
        else:
            k=1
        for j in range(k,len(new_data[i])):
            predicates.append(new_data[i][j].strip('][').split(', ') [1])
    predicates = np.array(predicates)
    predicates = np.unique(predicates)
    predicates= predicates.tolist()[1:]
    for i in range(len(predicates)):
        predicates[i] = [predicates[i].strip(" ' ")]
    return data+predicates
```

```python
original_file = "../../data/ConvAI2/train_both_original.txt"
final_file = "../../data/ConvAI2/Final_files/train_both_original_final.txt"
train_data = generate_data(original_file,final_file)
```

# preparing trianing data

```python
import numpy as np

def filtering(input_file,output_file):
    data = []
    keys = []
    with open(input_file, "r") as f:
            word_list = f.read().split("\n")
    for i in range(len(word_list)):
        data.append(word_list[i].split("\t"))
    for i in range(len(data)):
        if data[i][0].isnumeric():
            k =2
        else:
            k=1
        for j in range(k,len(data[i])):
            if data[i][j].strip('][').split(', ') [1]:
                if data[i][j].strip('][').split(', ') [1].strip(" ' ") != '<blank>':
                    keys.append(i)
    keys = list(set(keys))

    with open(input_file, "r") as f:
        lines = f.readlines()
        for i in range(len(lines)):
            if "partner's persona: " in lines[i]:
                lines[i] = lines[i].replace("partner's persona: ","")
            elif "your persona: " in lines[i]:
                lines[i] = lines[i].replace("your persona: ","")
            else:
                pass

    for i, line in enumerate(lines):
        if i in keys:
            output_file.write(lines[i])
```

# Generate vectors for each sentance

```python
import numpy as np
import tensorflow as tf
import ast
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
def encoding(input_file,simple_vec,word_vectors):
    word_list = []
    data = [] 
    question_answer = {}
    vectors_list = []
    gru = tf.keras.layers.GRU(100)
    with open(input_file, "r") as f:
            word_list = f.read().split("\n")
    for i in range(len(word_list)):
        data.append(word_list[i].split("\t"))
    for i in range(len(data)):
        if data[i][0].isnumeric():
            k =2
        else:
            k=1
        for j in range(k,len(data[i])):
            if data[i][j].strip('][').split(', ') [1].strip(" ' ") != '<blank>':
                for x in range(len(tokenizer.tokenize(data[i][k-1]))):
                               vectors_list.extend(word_vectors[tokenizer.tokenize(data[i][k-1])[x]])
                arr = np.array(ast.literal_eval(str(vectors_list).strip()))
                arr.shape= [1,len(arr),1]
                output = gru(arr)
                output = output.numpy().tolist()
                question_answer[i] = {'sentance':[(v+1) * 10 for v in output[0]],'predicate': [(v+1) * 10 for v in predicate_word_vectors[data[i][j].strip('][').split(', ') [1].strip(" ' ")].tolist()]}
            simple_vec.write(str(question_answer) +"\n")               
            vectors_list.clear()
            question_answer.clear()
    simple_vec.close()
```
