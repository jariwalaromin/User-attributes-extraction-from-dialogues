```python
# Python program to generate word vectors using Word2Vec 

# importing all necessary modules 
from nltk.tokenize import sent_tokenize, word_tokenize 
import warnings 

warnings.filterwarnings(action = 'ignore') 

import gensim 
from gensim.models import Word2Vec 
from gensim.test.utils import get_tmpfile
from gensim.models import KeyedVectors
import numpy as np
```

```python
from nltk.tokenize import RegexpTokenizer
word_list = []
data = [] 
predicates = []
input_file = "../../data/ConvAI2/Final_files/train_both_original_final.txt"
with open(input_file, "r") as f:
        word_list = f.read().split("\n")
for i in range(len(word_list)):
    data.append(word_list[i].split("\t"))

for i in range(len(data)):
    if data[i][0].isnumeric():
        k =2
    else:
        k=1
    for j in range(k,len(data[i])):
        predicates.append(data[i][j].strip('][').split(', ') [1])
predicates = np.array(predicates)
predicates = np.unique(predicates)
predicates= predicates.tolist()[1:]
for i in range(len(predicates)):
    predicates[i] = [predicates[i].strip(" ' ")]
predicates
```

```python
# Create CBOW model 
model1 = gensim.models.Word2Vec(predicates, min_count = 1, 
                            size = 100, window = 5) 
#save model
word_vectors = model1.wv
word2vecmodel = get_tmpfile("./Models/predicate_vectors.kv")
word_vectors.save(word2vecmodel)
```

# Generate vectors for each sentance

```python
vectors_list=[]
final_dict = {}
max_len = 0
f = open("./Encoded_Data/Predicates_Vectors.txt", "a")
for i in range(len(predicates)):
    for j in range(len(predicates[i])):
        vectors_list.extend([[(i+1)*10 for i in word_vectors[predicates[i][j]].tolist()]])
final_dict[0]= vectors_list
f.write(str(final_dict) +"\n")
#vectors_list.clear()
f.close()
```
